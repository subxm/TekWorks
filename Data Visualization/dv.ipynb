## Import reqeuired libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Load the dataset

```python
df = pd.read_csv('dataset.csv')
```

```python
df
```

## 1. Histogram

```python
plt.figure(figsize=(10,6))  # set the figure size

sns.histplot(df['age'], kde=True, bins=30, color='red', edgecolor='black', alpha=0.7)
# kde = kernel density estimate is the continuous line that represents the distribution of the data
plt.title("Age Distribution")

plt.xlabel("Age")
plt.ylabel("Frequency")

plt.grid(True)
plt.xlim(0, 100)
plt.ylim(0, 40)

plt.show()
```
