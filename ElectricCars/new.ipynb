```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Find missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Find the outliers using a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(df['PriceEuro'])
plt.title('Box Plot of Price')
plt.ylabel('Price')
plt.show()
```

```python
# Convert price to indian rupees
df['PriceINR'] = df['PriceEuro'] * 106
# Find the outliers using a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(df['PriceINR'])
plt.title('Box Plot of Price in INR')
plt.ylabel('Price in INR')
plt.show()
```

```python
# Segregate the car types PowerTrain column and plot the distribution of car types pie chart
car_type_counts = df['PowerTrain'].value_counts()
plt.figure(figsize=(8, 8))
car_type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Car Types')
plt.ylabel('')
plt.show()
```

```python
# Bodystyle distribution other than bar graph
bodystyle_counts = df['BodyStyle'].value_counts()
plt.figure(figsize=(10, 6))
bodystyle_counts.plot(kind='bar')
plt.title('Distribution of Body Styles')
plt.xlabel('Body Style')
plt.ylabel('Count')
plt.show()
```
