## Step1: Import Required Library

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

## Step2: Get URL from Amazon

```python
URL = "https://www.amazon.in/s?k=Laptop&rh=n%3A1375424031"
```

```python
# Make header to mimic a browser visit
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}
```

```python
# Send a GET request to the URL
response = requests.get(URL, headers=headers)

response
```

```python
# beautifulsoup to parse URL content
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
# Product find div
products = soup.find_all("div", class_ = "sg-col-inner")
len(products)
```

```python
# Create empty lists to store data
product_names = []

# Loop through each product and extract the name
for product in products:
    # find the product title
    title_tag = product.h2
    if title_tag:
        product_name = title_tag.text.strip()
        product_names.append(product_name)



        # find the product price
        price_tag = product.find("span", class_="a-price-whole")
        if price_tag:
            price = price_tag.text.strip()
            print(f"Product Name: {product_name}, Price: {price}")
```
